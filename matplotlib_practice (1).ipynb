# Matplotlib

```python
import matplotlib.pyplot as plt

x=[1,2,3]
y=[3,2,5]

plt.scatter(x,y) #plotting points
plt.show()
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)  # joins scattered points
plt.title("My 1st Graph") 
plt.show() 
```

```python
x2=[1,2,3]
y2=[5,3,4]

plt.plot(x,y,label='line-1')
plt.plot(x2,y2,label='line-2')

plt.legend()  #distinguish between lines
plt.show()

```

# Bar Chart

```python
left=[1,2,3,4,5] #x-coordinates of left sides of bars
h=[10,20,5,15,17] # heights of bars 
l=['a','b','c','d','e'] #labels of bars

#plotting bar chart
plt.bar(left,h,tick_label=l,width=0.8,color={'red','blue','yellow'})
```

# Plotting Histogram

```python
ages=[2,5,6,3,4,10,15,10,16,25,3,46,55,13,96,75,86,64,35,53,12,94,65,56]

#setting ranges and intervals
range=(0,100)
bins=10

#ploting histogram 
plt.hist(ages,bins,range,color='blue',histtype='bar',rwidth=0.7)
```

# Pie chart

```python
activities=['eat','sleep','work','play']  #defining labels
slices=[3,8,6,2] #portion covered by each label
colors=['r','g','b','y']
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0,0.1,0),radius=1.5,autopct='%1.1f%%')
```

# Plotting Curves

```python
#plotting sine wave
import numpy as np
x=np.arange(0,2*(np.pi),0.1)
y=np.sin(x)

plt.ylim(-2,2)
plt.plot(x,y)
```

```python
y=np.cos(x)
plt.ylim(-2,2)
plt.plot(x,y)
```

# Plotting subplots

```python
def create_plot(ptype):
    x=np.arange(-10,10,0.01)  #x-axis values
    
    if(ptype=='linear'):
        y=x
    elif(ptype=='quadratic'):
        y=x**2
    elif(ptype=='cubic'):
        y=x**3
    plt.plot(x,y)
    
if __name__=="__main__":
    create_plot('linear')
    create_plot('quadratic')
    create_plot('cubic')
        
```
