```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=pd.read_csv('Training Data/Train.csv')
```

```python
x.head()
```

```python
x=x.values
```

```python
y=x[:,-1]
x=x[:,:-1]
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
#prediction
xtest=pd.read_csv('Testing Data/Test.csv')

output=model.predict(xtest)

bias= model.intercept_
coeff=model.coef_

model.score(x,y)
print("y = ",coeff,"x + ",bias)
output.shape
```

```python
#visualise 
xtest=xtest.values

plt.scatter(x[:,0],y,label='feature 1')
plt.scatter(x[:,1],y,label='feature 2')
plt.scatter(x[:,2],y,label='feature 3')
plt.scatter(x[:,3],y,label='feature 4')
plt.scatter(x[:,4],y,label='feature 5')

plt.plot(xtest[:,1],output,color='black',label='Prediction')
plt.legend()
plt.show()
```
